## Exercice 3 - Introduction à NumPy

NumPy (Numerical Python) est une puissante bibliothèque mathématique disponible en Python. Il s'agit d'une bibliothèque qui devient vite essentielle lorsque l'on aborde des sujets plus avancés en Python, comme l'apprentissage par machine (machine learning). NumPy est très rapide, notamment grâce à une implémentation de son noyau dans le langage de programmation C.

On peut créer un tableau comportant 5 zéros de la manière suivante :

```python
import numpy as np

tableau = np.zeros(5)
print(tableau)
```

De la même manière, on peut créer un tableau rempli de 1 :

```python
tableau = np.ones(10)
print(tableau)
```

```python
depart = 1
arrivee = 1000
nb_elements = 5
echelle_lineaire = np.linspace(depart, arrivee, nb_elements)
print(echelle_lineaire)
```

Il est également possible de créer un tableau NumPy à partir d'une liste Python

```python
liste = [1, 2, 3, 4, 5]
tableau = np.array(liste)
print(tableau)
```

A ce stade, il est intéressant de regarder les dimensions des tableaux NumPy

```python
tableau = np.array([1, 2, 3])
dimension = tableau.shape
print(dimension)
```

```python
tableau = np.array([[1, 2, 3], [4, 5, 6]])
print(tableau)
dimensions = tableau.shape
print(f"dimensions = {dimensions}")
```

## Exercice 5 - Introduction au traitement d'image avec NumPy

Une image peut être vue comme un tableau à 2 dimensions de pixels (picture elements). Chaque pixel d'une image est caractérisé par une couleur.

Il existe différentes manières de représenter une couleur. La plus commune est la représentation par 3 nombres : Rouge, Vert, Bleu. En anglais, cela donne Red, Green Blue, soit RGB. Une couleur au format RGB peut être représentée par un vecteur dont chaque composante est comprise entre 0 et 255 :

$$
\text{Couleur} =
\begin{pmatrix}
    R \\
    G \\
    B
\end{pmatrix},
\quad\text{Rouge} =
\begin{pmatrix}
    255 \\
    0 \\
    0
\end{pmatrix},
\quad\text{Vert} =
\begin{pmatrix}
    0 \\
    255 \\
    0
\end{pmatrix},
\quad\text{Bleu} =
\begin{pmatrix}
    0 \\
    0 \\
    255
\end{pmatrix},
\quad\text{Blanc} =
\begin{pmatrix}
    255 \\
    255 \\
    255
\end{pmatrix},
\quad\text{Noir} =
\begin{pmatrix}
    0 \\
    0 \\
    0
\end{pmatrix}.
$$

Si on considère qu'une couleur est une dimension, alors une image comporte donc 3 dimensions : sa hauteur, sa largeur et ses couleurs.

Nous avons vu que NumPy était très efficace pour manipuler des tableaux. Comme une image est un tableau à 3 dimensions, NumPy peut les manipuler.

Il est possible de réduire une dimension en passant de la couleur aux niveaux de gris. Ainsi, au lieu d'utiliser 3 composantes RGB pour chaque pixel, on utilise une unique composante. Cette composante, entre 0 et 255, représente le niveau de gris.

Cette transformation peut s'effectuer de manière très simple :

```python
from skimage import color
sapin_gris = color.rgb2gray(sapin)
plt.imshow(sapin_gris, cmap="gray")
plt.colorbar()
```

On observe que seule la luminescence est conservée. Les valeurs de pixels sont ici normalisés entre 0 et 1. Il suffirait de multiplier par 255 pour obtenir l'intervalle classique $[0;255]$.

```python
from skimage.exposure import histogram
H = histogram(sapin_gris)
plt.plot(H[0], label="Fréquence des niveaux de gris")
plt.legend()
```

## Exercice 6 - Traitement du signal

En mathématiques, on considère qu'**une image est un signal**. Le **traitement du signal** est une branche des mathématiques qui s'applique aussi bien au traitement d'images, de sons, signaux électriques, etc.

Un **signal** est une mesure d'une grandeur physique quelconque qui porte une information. Le but du traitement du signal est d'établir des représentations et des manipulations mathématiques afin de déterminer des relations entre ces représentations et effectuer des modifications. Pour cela, on utilise un modèle mathématique qui est une fonction à 1, 2, ou 3 variables.

Ainsi, le son peut être représenté en fonction du temps, donc avec une fonction à 1 variable. Une image peut être représentée en fonction de 2 variables : abscisse et ordonnée d'un pixel.

La **convolution** est une opération importante en traitement du signal. Si on prend 2 signaux $f$ et $g$ à 1 variable continue $t$, alors on note $(f * g)(t)$ la convolution de f et g et se définie comme suit :

$$
\forall t \in \mathbb{R}^{+}, (f * g)(t) = \int_{- \infty}^\infty f(u) g(t - u) du
$$

Dans le cas d'une image, qui comporte 2 variables discrètes que l'on notera $x$ et $y$, il s'agit d'une opération entre 2 images en niveaux de gris $I_1$ et $I_2$, notée $(I_1 * I_2)(x, y)$ telle que :

$$
\forall (x, y) \in \mathbb{N}^{2}, (I_1 * I_2)(x, y) = 
    \sum_{i = - \infty}^{\infty} \sum_{j = - \infty}^{\infty} I_1(i, j) I_2(x - i, y - j)
$$

Ces 2 définitions de la convolution sont équivalentes. Dans le 1er cas, il s'agit de la définition dans un espace temporel qui est de nature continue, avec 1 variable. Dans le 2e cas, il s'agit de la définition dans un espace discret, avec 2 variables.

En pratique, le filtre de flou Gaussien peut être implémenté en appliquant une convolution à une image. La convolution est également utilisée pour l'**extraction des contours** d'une image. On va s'intéresser à l'extration des contours dans cet exercice.
